```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

1. (a) 

$p(y) = 0.5N(1, \sigma^2) + 0.5N(2, \sigma^2)$

```python
mu1 = 1
mu2 = 2
sigma = 2
```

```python
# Green line is the marginal probability density for y
x = np.linspace(mu1 - 3*sigma, mu2 + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu1, sigma), '--')
plt.plot(x, stats.norm.pdf(x, mu2, sigma), '--')
plt.plot(x, 0.5*stats.norm.pdf(x, mu1, sigma)+0.5*stats.norm.pdf(x, mu2, sigma))
plt.show()
```
